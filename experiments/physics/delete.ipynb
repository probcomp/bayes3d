```python
import bayes3d as b
```

```python
rend = b.MJFC_Physics_Renderer("basic")
rend.add_light([0,0,1])
rend.add_floor([2,2,2])
rend.add_object('box', 'box', pos = [0.2,0,0.02], size="0.02 0.02 0.02", rgba=[1,0,0,1] )
rend.add_object('sphere', 'ball', pos = [-0.2,0,0.02], size="0.02", rgba=[0,1,0,0.5] )
rend.set_object_pose('box', qpos=[0.01,0,0.15,1,0,0,0], keyframe=0)
# rend.set_object_velocity('box', qvel=[-0.2,0,0,0,0,0], keyframe=0)
rend.render_image()
print(rend.get_xml())
```

```python
rend.render_video(4)
```
